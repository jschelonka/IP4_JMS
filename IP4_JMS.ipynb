```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
death_cause = pd.read_csv('Death Cause Reason by Country.csv')
```

```python
death_cause.info()
```

```python
death_cause.describe()
```

# Heat Map Comparison of Causes of Death

```python
heat_map = death_cause[['Country Name', 'Covid-19 Deaths', 'Cardiovascular diseases', 'Kidney diseases', 'Interpersonal violence', 'HIV/AIDS', 'Tuberculosis', 'Maternal disorders', 'Lower respiratory infections', 'Alcohol use disorders', 'Diarrheal diseases', 'Poisoning', 'Nutritional deficiencies']].corr(method = 'pearson')
cols = ['Country Name', 'Covid-19 Deaths', 'Cardiovascular diseases', 'Kidney diseases', 'Interpersonal violence', 'HIV/AIDS', 'Tuberculosis', 'Maternal disorders', 'Lower respiratory infections', 'Alcohol use disorders', 'Diarrheal diseases', 'Poisoning', 'Nutritional deficiencies']
axis = sns.heatmap(heat_map, annot = True,
                    xticklabels = cols,
                    yticklabels = cols,
                  cmap = 'BuPu')
```

# Scatter plot comparing Country Name and Covid-19 Deaths
